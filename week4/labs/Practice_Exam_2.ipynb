##### 3.1 Solution

````python
# @title
# Time complexity: O(n^2)

# This is a nested loop where the number of iterations of the inner loop depends on the current iteration of the outer loop. In the worst case scenario,
# it's roughly equivalent to a full nested loop, which gives us a time complexity of O(n^2).
# The number of operations will be n + (n-1) + (n-2) + (n-3) + ... + 1 (where n is the length of the list). This is approximately equal to n(n+1)/2 (can you figure out why?), which is O(n^2).
````
