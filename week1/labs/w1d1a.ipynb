## Question 4: Putting it all together!

A nice way to do operations on a variable is to use shorthand operators!

Here's one way to calculate $(10 * 10 + 20) / 5$ step by step

````python
# Lets start with 10
answer = 10

# Multiply that by another 10
answer = answer * 10

# Add 20 to it
answer = answer + 20

# Then divide it all by 5
answer = answer / 5

print(answer)
````

Notice how I kept doing `answer = answer * ...` ?

Instead python lets us do `answer *= ...`, `answer += ...`, and `answer /= ...` !

These are called _short_hand operators because when we type `answer *= x` we're telling the computer to actually do `answer = answer * x `

Like:

````python
# Lets start with 10
answer = 10

# Multiply the answer by another 10
# answer = answer * 10
answer *= 10

# Add 20 to the answer
# answer = answer + 20
answer += 20

# Then divide the answer by 5
# answer = answer / 5
answer /= 5

# And we get the same answer!
print(answer)
````
