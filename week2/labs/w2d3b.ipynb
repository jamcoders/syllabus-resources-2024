One of the most important things in biology is DNA, it encodes __everything__ about living things. DNA is made of a pair of 2 long sequences composed of 4 different *bases*: Adenine (A), Thymine (T), Cytosine (C), and Guanine (G). Each *base* fits like a puzzle piece to pair up with **a specific** pairing base in the other sequence.

**A**denine pairs up with **T**hymine (**AT** or **TA**  base pair) and **C**ytosine pairs up with **G**uanine (**CG** or **GC** base pair). This pairing is important because DNA is double stranded. Each DNA strand is made up of multiple bases and fits with its other strand using the puzzle piece pairing rules. If we know the bases that make up one strand of DNA, we can figure out which bases make up the other strand by finding each base's complement (the other piece of the puzzle pair). For example:

- `"AGATCC"` -- This is the strand we know.
    - We know that our strand has 6 bases, so our pairing strand will have 6 bases
    - We know that every base in the pairing strand will have to make a pair with the first strand
    - We know that each base in the strand will pair to **a specific** pairing strand.
- The pairing strand must be: `"TCTAGG"`
    - This is because the pairing strand must be 6 bases long AND:
    - The first strand's `A` pairs with our `T`  (**AT** base pair)
    - The first strand's `G` pairs with our `C`  (**GC** base pair)
    - The first strand's `A` pairs with our `T`  (**AT** base pair)
    - The first strand's `T` pairs with our `A`  (**TA** base pair)
    - The first strand's `C` pairs with our `G`  (**CG** base pair)
    - The first strand's `C` pairs with our `G`  (**CG** base pair)
