## Challenge 1: Ancient Egyptian multiplication method
The ancient Egyptians had a method of multiplying integers using only addition, doubling, and halving. Here is how the method works:

1. Take two numbers to be multiplied and write them down at the top of two columns.
2. In the left-hand column repeatedly halve the previous number, discarding any remainders, and write the result below the last in the same column, until you write a value of 1.
3. In the right-hand column repeatedly double the previous number and write the result below. stop when you add a result in the same row as where the left hand column shows 1.
4. Examine the table produced and discard any row where the value in the left column is even.
5. Sum the values in the right-hand column that remain to produce the result of multiplying the original two numbers together

**For example**:   17 × 34
Halving the first column:<br>

17 | 34
--|--
8 |
4 |
2 |
1 |

Doubling the second column:

17 | 34
-----|-----
8 | 68
4 | 136
2 | 272
1 | 544

Strike-out rows whose first cell is even:

17 | 34
---- | -----
8|<s>68</s>
4|<s>136</s>
2|<s>272</s>
1|&emsp;544

Sum the remaining numbers in the right-hand column:

17 | 34
---- | ----
8 | --
4 | ---
2 | ---
1 | 544
**Total** | 578

So 17 multiplied by 34, by the Egyptian method is 578.
